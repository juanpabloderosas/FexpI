### Probabilidad y Física

La teorı́a de probabilidad en ciencias empı́ricas –y en particular en fı́sica– juega muchos papeles. En FExp, la teorı́a de probabilidad se utiliza para modelar datos experimentales, que ulteriormente se compararán con modelos teóricos. La forma de las comparaciones puede ser muy diversa, por lo que el interjuego de los modelos de datos con los modelos teóricos es rico en ambigüedades, y suele sorprender o enojar a auellos que poseen una imágen monolı́tica de la ciencia, la cual generalmente es encontrada en los libros de divulgación.

### Variable aleatoria:

Una variable aleatoria $X$ transforma los posibles resultados de un experimento (operación de medida) a números reales ó enteros.

### Muestreo con Reemplazo
Cuando cada una de las $X_i$ tiene la misma distribución de probabilidad. En esta situación, diremos
que muestreamos ***variables aleatorias Independientes e Idénticamente Distribuídas (IID).***

### Muestreo sin Reemplazo

La realización de un experimento modifica la distribución de probabilidad de una repetición
posterior del experimento. Si al primer experimento lo modelamos con la variable aleatoria $X_1$ (que tiene una distribución de probabilidades), entonces la repetición del experimento, que modelamos por $X_2$ , no tendrá la misma distribución de probabilidad,..., ası́ hasta $X_n$ .

## Probabilidad a partir de la frecuencia relativa:

#### Definición (ojo que está escueta)

Si un experimento se repite $n$ veces bajo condiciones idénticas y $n_x$ es el número de veces que el resultado del experimento es $X = x$, entonces:

\begin{equation}
p(x) = \lim_{n \rightarrow \infty} \dfrac{n_x}{n} 
\end{equation}

Los resultados de un experimento pueden ser \textit{discretos} (como contar partículas en un detector) o \textit{continuos} (como medir la energía de un sistema), y nuestros resultados serán modelados como variables aleatorias. Cada variable aleatoria tendrá un modelo de probabilidad, llamado habitualmente \textit{distribución de probabilidad}. Una distribución de probabilidad puede ser descrita convenientemente por dos funciones $p(x)$ y $F(x)$, cuyas definiciones damos a continuación para variables continuas.

#### PDF ó FDP: Probability Density Function ó Función Densidad de Probabilidad

Para las variables continuas se define una \textit{función densidad de probabilidad} $f(x)$. La probabilidad de que $X$ esté en el intervalo $[x = a,x = b]$ está dada por la integral de $f(x)$ en ese intervalo.

Para $f(x)$ tenemos:

$$
p\lbrace a \leq X \leq b \rbrace  = \int_{a}^{b} f(x) dx \\
f(x) > 0 \;\; \forall \; \; -\infty < x < \infty \\
\int_{-\infty}^{\infty} f(x) dx = 1$$


Función contínua de densidad definida como $f(X = x) = \lim_{\Delta x \rightarrow 0} \; \dfrac{f(x < X < x+\Delta x)}{\Delta x} $, aunque más formalmente sería mejor $f(x) = \dfrac{dF(x)}{dx}$.

#### FPA: función probabilidad acumulada

En el caso continuo la probabilidad acumulada está dada por la \textit{función de densidad de probabilidad acumulada}, $F(x)$:


$$
F(x) = p\lbrace X \leq x \rbrace = \int_{-\infty}^{x} f(x) dx \\
F(-\infty) = 0 ;  F(\infty) = 1 \\
p(a < X <b) = F(b) - F(a) \\
f(x) = dF(x)/dx
$$
Formalmente se define probabilidad acumulada $F(x)$ y la función densidad de probabilidad sale de aquí.


Con esto tenemos definidas las funciones de las cuales sale toda la información de un modelo aleatorio continuo. Es eso ó ***la Función Generadora de Momentos*** (la dejamos para después).

## Propiedades *descriptivas* de una distribución

### Momentos de orden $r$

\begin{equation}\label{emom}
\mu '_r = \langle X^r \rangle = 
\begin{cases}
    \sum_{x} x^r p(x) & \text{(caso discreto)} \\
    \int_{-\infty}^{\infty} x^r f(x) dx & \text{caso continuo}
  \end{cases}
\end{equation}

### El primer momento es la ***media*** de la distribución

$$\mu'_{r = 1} =  \mu = \langle X \rangle =  \int_{-\infty}^{\infty} x f(x) dx $$

### Momentos *centrados*  en la media de orden $r$

\begin{equation}\label{emomcen}
\mu_r = \langle X^r - \mu \rangle = 
\begin{cases}
    \sum_{x} (x - \mu)^r p(x) & \text{(caso discreto)} \\
    \int_{-\infty}^{\infty} (x-\mu)^r f(x) dx & \text{(caso continuo)}
  \end{cases}
\end{equation}

### Momento centrado de orden 2 ó la Varianza de $x$

$$ Var(X) = \sigma^2_X = \langle (X - \mu)^2 \rangle = \int_{-\infty}^{\infty} (x - \mu)^2 f(x) dx $$


**La desviación estándar (una típica medida de dispersión), es la raíz cuadrada de la varianza:**

$$\sigma_X =  \sqrt{(Var(X)}$$

### Distribución Normal

La distribución normal (en física todes decimos gaussiana, pero no es de uso fuera de la disciplina) es la más importante y ampliamente usada de las distribuciones de probabilidad. La razón de esto es que reside en el corazón de la estadística, con un teorema muy importante llamado \textit{Teorema del Límite Central} o TLC (No confundir con Tratado de Libre Comercio). Este teorema, que enunciaremos formalmente al final de este capítulo, dice que \textit{la distribución de medias muestrales}, $\overline{X} = \sum X_i /N$, \textit{casi sin importar la distribución de} $X_i$, \textit{es una distribución normal}.

Por último, la distribución normal es tan \textit{normal}, que hay una forma específica de decir que una variable aleatoria tiene esta distribución. Si  $X$ pertenece a una distribución normal de media $\mu$ y de desviación estándar $\sigma$, decimos que $X \approx \mathcal{N}(\mu,\sigma^2)$.

#### Distribución Normal: Función Densidad de Probabilidad


\begin{equation}\label{efnorm}
f(x) = \dfrac{1}{\sqrt{2\pi}\sigma} exp \Bigg( -\dfrac{(x-\mu)^2}{2\sigma ^2} \Bigg)
\end{equation}
\begin{equation} \nonumber
\text{para } -\infty < x <  \infty \; ; \; -\infty < \mu < \infty \; ; \; 0 < \sigma^2 < \infty
\end{equation}

donde $f(x)$ en la ec.\eqref{efnorm} está parametrizada con la media $\mu$ y la desviación estándar $\sigma$.

#### Distribución Normal: Función Densidad de Probabilidad, en código
Esta función está en el comando `dnorm(x,mean,sd)` pueden saber los valores de la función, que están en el paquete base de R.

\begin{equation}\label{eFnorm}
F(x) = p(X \leq x) = \dfrac{1}{\sqrt{2\pi}\; \sigma}\int_{-\infty}^x exp \Bigg( -\dfrac{(t-\mu)^2}{2\sigma^2} \Bigg) dt

# Muestras y distribución normal

Un problema: la definición de probabilidad requiere ***infinitas*** medidas. El conjunto hipotético de todas las medidas ***levantaría*** una distribución, en este caso normal. A ese conjunto hipotético se le denomina ***población***.

El problema, siempre, es intentar ***morderse la cola, a la manera de la serpiente que da origen al infinito***: las muestras son conjuntos finitos de datos, y las cantidades como la media (promedio es la palabra popular) pertenece a un conjunto infinito. No nos molestamos más: existen versiones que apodamos con ***muestral*** para $\mu$ y $\sigma$.

Una muestra es un cierto ***conjunto finito de datos***. Vamos a comenzar con las medidas de g del acelerómetro de la tarea (su próximo laboratorio).

```r
setwd("/home/juan/Documentos/Docencia/fexpi/2019/Laboratorios/Laboratorio4/Datos/") #indica un nuevo directorio

g <- read.csv(file = "00.tsv", sep = "\t", header = F)

g <- g[ , c(-1,-2,-6)]; colnames(g) <- c("gx","gy","gz");

g$g.mod <- sqrt(g$gx^2 + g$gy^2 +g$gz^2)

head(g, 3) #saca primer columna
```

```r
hist(g$g.mod, breaks = nclass.FD(g$g.mod), col = "Deepskyblue4", main = "Medidas de g con acelerómetro", xlab = expression(g(m/s^2)), ylab = "Frecuencia") 
```

#### Media muestral

Se puede comparar con la definición de la media de la *población*, es igual, pero en lugar de una integral (no sabemos la FDP), tenemos una ***sumatoria de datos***. La media muestral de nuestra medida será:

$$
\overline{g} = \frac{1}{N} \sum_{i = 1}^N g_i
$$

Podemos calcularla para los datos *limpios y sucios*:

```r
media.g.limpia <- mean(g.limpia); media.g.sucia <- mean(g.sucia);
sprintf("La media de los datos limpios es %f", media.g.limpia)
sprintf("La media de los datos sucios es %f", media.g.sucia)
```

Casi no difieren porque son demasiados datos (el tamaño de la muestra es grande, alrededor de N = 5000)

### Desviación estándar de la muestra

La desviación estándar mide el ancho de la función densidad de probabilidad. Obvio, sin medidas infinitas hay que usar la ***desviación estándar de la muestra***, $s$:
$$
s = \sqrt{ \sum_{i = 1}^{N} \frac{(g_i - \overline{g})^2}{N-1} }
$$

como siempre, es bueno comparar con la definición de la integral:
* s es para un número finito de datos.
* la integral pasa a ser sumatoria...pero hay que ver cómo aparecen estas cuentas...

#### Desviación estándar de la muestra

Otro problema de $s$ es que no es ***robusta*** frente a ***suciedades*** o datos incorrectos:

* como hay un cuadrado en la definición, los valores de $s$ cambian mucho cuando hay ***outliers*** que están lejos de la media.

* si queremos que $s$ se parezca a $\sigma$ de la distribución normal, entonces es mejor que usemos `g.limpia` para calcular $s$

```r
sprintf("s_sucia = %f", sd(g.sucia))
sprintf("s_limpia = %f", sd(g.limpia))
```

Vemos que:
* `s.sucia` vale más del doble que `s.limpia`
* Es bueno sacar los ruidos: hay maneras teóricas, pero por ahora no vamos a ver sino estas maneras brutales.
* Como $s$ aproxima $\sigma$, y $\sigma$ cuantifica probabilidades, es importante conocer el valor de $s$, ya que nos puede sobreestimar grandemente el ***error de una medida***.
