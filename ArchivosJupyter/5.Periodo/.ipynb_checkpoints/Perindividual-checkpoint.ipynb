## 3. Modificadores para read.csv()

**sep = ":"** va a separar columnas con los dos puntos

**skip = 2** saca "Total Time..." y los nombres de la columnas (más cómodo)

**header = F** como salteamos el nombre de las columnas, no le ponemos encabezado

```r
p <- read.csv("periodos.csv", sep = ":", skip = 2, header = F)
```

```r
head(p,5); tail(p,5)
```

```r
p <- p$V4; hist(p)
```

## Limpiamos datos

Sacamos *valores extremos*

```r
p.limpio <- p[p < (mean(p) + 0.35) &  p > (mean(p) - 0.35)  ]
```

```r
length(p); length(p.limpio)
```

```r
hist(p.limpio, breaks = nclass.FD(p.limpio))
library(fitdistrplus)
```

```r
a <- fitdistr(p.limpio, densfun = "normal")
hist(p.limpio, breaks = nclass.FD(p.limpio))
a
```

```r
llplot(a, expansion = 2)
```
