# Ejemplos de Algoritmos

### 1. Ángulo entre vectores (un ejemplo simple )

Como ejemplo simple ponemos el cálculo del ángulo entre dos vectores. Un diagrama de flujo es un gráfico que conecta "bloques" de un código.

Por ejemplo, ¿qué necesitamos hacer para calcular el ángulo entre dos vectores?

Sabemos que, dados dos vectores $\vec{a}$ y $\vec{b}$, el ángulo $\alpha$ entre ellos estará dado por:

$$
\alpha = cos^{-1}\Bigg(\frac{\vec{a} \cdot \vec{b} } {|\vec{a}| |\vec{b}|}  \Bigg)
$$

Entonces, para saber el ángulo $\alpha$, necesitamos:

1. **`Dos vectores de entrada (Input de nuestro código)`** $\Rightarrow$  2. **`Calcular el producto punto entre los vectores`** $\Rightarrow$ 3. **`Calcular el módulo de cada vector`**$\Rightarrow$4. **`Dividir el producto punto por el producto de los módulos`**$\Rightarrow$ 5. **`Calcular al arco coseno del resultado anterior`** $\Rightarrow$ 6. **`Expresar el resultado en radianes y grados sexagesimales`**

```r
# 1. Dos vectores de entrada
a <- c(1,0,1); b <- c(1,0,-1)
#2. Calcular el producto punto
prodpunto <- sum(a*b)
#3. Modulos
mod.a <- sqrt(sum(a^2)); mod.b <- sqrt(sum(b^2));
#4. Dividir prodpunto por (mod.a * mod.b)
p <- prodpunto/(mod.a*mod.b)
#5. calcular el arco coseno de p
alpha <- acos(p)
#6. Expresar el resultado en radianes y grados sexagesimales
cat(bold$blue("El ángulo entre a  y b  es de", alpha, "rad"))
cat(bold$red("\nEl ángulo entre a y b es de", alpha*180/pi, "°"))
```

### 2. Superficie plana de una mano.

1. Saco una foto de mi mano sobre una hoja A4 (29.7 x 21 cm)
2. Usamos el contraste (distinguimos negro de blanco) para seleccionar píxeles.
3. Sabiendo el número de píxeles de la mano respecto del total, calculamos la superficie plana de mi mano.

Esto se ve más o menos como sigue:

```r
####ESPECIFICAR CARPETA DE TRABAJO
setwd("/home/juan/Documentos/Docencia/fexpi/2020/NTB/1.rintro/tex1/Archivos/")
#leemos el archivo de escala de grises (0 = negro/ 1 = blanco/ 0< grises <1)
#Si no está en enscala de grises no va a funcionar...
mano <- readJPEG('mano.jpg') #cargamos la imagen
mano <- as.matrix(mano[,,1]) #la hacemos una matriz
```

Una foto en escala de grises es una ***matriz*** de números (0 = negro,0 < grises < 1, 1 = blanco) que forman la imagen ***digital***:

```r
mano[125:130, 85:95] #los más chiquitos, más cerca del negro, parecidos a uno, cerca del blanco
```

```r
#### Ploteo de imágenes#############
fig(width = 9, heigth = 12);plot(c(1,ncol(mano)) , c(1,nrow(mano)) , type = "n", main = "Mano", xlab = "", ylab = "" )
rasterImage(image = mano, xleft = 1, ybottom = 1, xright = ncol(mano), ytop = nrow(mano))
show(range(mano))
```

Elegimos un valor de contraste para ***seleccionar*** los píxeles de la mano, que son más oscuritos que el resto. Para esto vemos un lindo histograma:

```r
fig(width = 22, heigth = 14)
hist.orig <- hist(mano, main = "Histograma pixeles (antes de su conversión)",
breaks = (b <- seq(-0.025,1.025,0.05)), col = grey.colors(length(b)-1, start = 0,end = 1), xlab = "(Negro = 0) <=====     Gris     =====> (Blanco = 1)",cex.axis = 1.25 ,cex.lab = 2 , plot = T ) 
```

```r
manosaturada <- !(mano < 0.65) #establecemos un límite de gris para la mano
storage.mode(manosaturada) <- "numeric" #pasa de logico a numerico T = 1, F = 0
hist(manosaturada, breaks = seq(-0.025,1.025, 0.05), col = c("Black","White" ,"White"), xlab = "Negro ó Blanco", cex.axis = 1.5, cex.lab = 2 ,main = "Después de la polarización política", probability = F)
```

```r
manosaturada <- !(mano < 0.65) #establecemos un límite de gris para la mano
storage.mode(manosaturada) <- "numeric" #pasa de logico a numerico T = 1, F = 0
####PLOTEO de SELECCIóN
#si no vemos todo el contorno negro, sería mejor aumentar el contraste
fig(width = 14, heigth = 20);plot(c(1,ncol(manosaturada)) , c(1,nrow(manosaturada)), type = "n", main = "Mano Saturada", xlab = "", ylab = 0)
rasterImage(image = manosaturada, xleft = 1, ybottom = 1, xright = ncol(manosaturada),
            ytop = nrow(manosaturada))
```

```r
pixmano <- sum(!manosaturada)#pixeles de la mano
pixfoto <- ncol(mano)*nrow(mano) #todos los pixeles
supmano <- (pixmano/pixfoto)*29.7*21 #regla de tres para la superficie de la mano...
#imprime resultados
cat(bold(red(sprintf("La superficie de la mano es %.1f cm²
\n\nLa superficie de una hoja A4 es %.1f cm² \n", as.numeric(supmano), 29.7*21))))
```
