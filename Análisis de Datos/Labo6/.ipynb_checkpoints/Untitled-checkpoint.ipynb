# Corrigiendo cosas: el eje del tiempo

Ya que vamos a comparar series temporales, tenemos que poner $t = 0$ igual para las dos series temporales:

* Para cada serie temporal:
        
     * Tomamos el primer valor de tiempo $t_1$.
     * A toda la serie temporal le restamos este valor: $t_{nuevo,i} \;=  t_i - t_1$
     
Haciendo esto, tenemos:     

```r
da$t <- da$t - da$t[1];
dsa$t <- dsa$t - dsa$t[1];
```

```r
gtT1 <- ggplot() + 
geom_point(data = da,aes(x = t, y = T, colour = "Pava abrigada")) + 
geom_point(data = dsa, aes(x = t, y = T, colour = "Pava desnuda")) + 
scale_colour_manual(values = c("Pava abrigada" = "Red3", "Pava desnuda" = "DeepSkyBlue4"), name = NULL) + 
temajuan + theme(legend.position = c(0.8,0.8)) +
scale_x_continuous(limits = c(-0.5, max(dsa$t)+0.5) ) + 
xlab(TeX("$t\\; (min)$")) + 
ylab(TeX("$T\\; (C)$")) 
```

# Analizando cosas

El modelo de enfriamiento de Newton es:

$$
T(t) = T_{amb} + (T_0 - T_{amb}\;)\; e^{-rt}
$$

* $t [s]$: es el tiempo.
* $T_0 \; [C]$: Temperatura incial del agua.
* $T_{amb} \; [C]$: Temperatura ambiente.
* $T(t)\; [C]$: Es la temperatura del agua en cada instante.

Además:

* $r \; [1/s]$: es el coeficiente de *enfriamiento* (o de intercambio de calor) del sistema ***"pava + agua"*** ó ***"pava + agua + abrigo"***

Este coeficiente ¿qué define? ¿Cómo se ven las gráficas para todos los parámetros constantes y diferentes $\mathbf{r}$?


```r
gtT <- ggplot() + 
geom_point(data = da,aes(x = t, y = T, colour = "Pava abrigada")) + 
geom_point(data = dsa, aes(x = t, y = T, colour = "Pava desnuda")) + 
scale_colour_manual(values = c("Pava abrigada" = "Red3", "Pava desnuda" = "DeepSkyBlue4"), name = NULL) + 
temajuan + theme(legend.position = c(0.8,0.8)) +
xlab(TeX("$t\\; (min)$")) + 
ylab(TeX("$T\\; (C)$")) 
```

```r
show(gtT)
```

# ¿Ya estamos conformes conque r caracteriza si el abrigo funciona?

Para determiar $r[s^{-1}]$ hay que hacer un par de cuentas con los datos.

$$
ln\Bigg( \frac{T(t) - T_{amb}}{T_0 - T_{amb}}  \Bigg) = -rt
$$

```r
#Temperaturas ambiente

Tad <- 34.0;
Taa <-24.0;

da$V <- (da$T - Taa)/(da$T[1] - Taa)
dsa$V <- (dsa$T - Tad)/(dsa$T[1] - Tad)
```

```r
gtT <- ggplot() + 
geom_point(data = da,aes(x = t, y = V, colour = "Medidas Pava abrigada"), alpha = 0.04, size = 6) + 
geom_point(data = dsa, aes(x = t, y = V, colour = "Medidas Pava desnuda"), alpha = 0.04, size = 6) + 
scale_colour_manual(values = c("Medidas Pava abrigada" = "Red2", "Medidas Pava desnuda" = "DeepSkyBlue4"), name = NULL,
                   guide = guide_legend(override.aes = list(size = 6, alpha = 0.6) ) ) + 


geom_smooth(data = dsa,aes(x = t, y = V, lty = "Ajuste Pava desnuda"), colour = "Magenta", lwd = 1.5, method = "lm", formula = "y ~ x", se = F, fullrange = T) + 
geom_smooth(data = da, aes(x = t, y = V, lty =  "Ajuste Pava abrigada"), colour = "Black", method = "lm", formula = "y ~ x", se = F, fullrange = T) + 
scale_linetype_manual(values = c("Ajuste Pava desnuda" = 1, "Ajuste Pava abrigada" = 1), name = NULL,
                     guide = guide_legend(override.aes = list(colour = c("Black", "Magenta")) )) + 

temajuan + theme(legend.position = c(0.8,0.8)) +
xlab(TeX("$t\\; (min)$")) + 
ylab(TeX("$\\frac{T - T_a}{T_o - T_a}\\; \\[escala \\; log\\]$")) + 
scale_y_continuous(trans = "log")
```

```r
show(gtT)
```

```r
abrigada <- summary.lm(lm(log(da$V) ~ da$t))
desnuda <- summary.lm(lm(log(dsa$V) ~ dsa$t))
cat(bold(red("Pava abrigada\n")))
abrigada$coefficients
cat(bold(blue("Pava desnuda\n")))
desnuda$coefficients

```
