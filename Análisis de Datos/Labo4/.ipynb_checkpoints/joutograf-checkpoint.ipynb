```r
d <- read.csv(file = "Esferas.csv", header = T, sep = ",")
```

```r
colnames(d) <- c("m", "d", "Nombre"); #masa en g, diámetros en cm
head(d, 2)
```

```r
#defino radios
d$r <- d$d /2;

#defino volumen y densidad volumetrica
d$V <- (4*pi/3)*d$r^3; #cm^3
d$rho <- d$m / d$V; #g/cm^3
head(d, 3)
```

```r
#ploteamos mal
gmal <- ggplot(data = d, aes(x = d, y = m)) + geom_point() + temajuan;
```

```r
#mejoramos
gejes <- gmal + xlab("d (cm)") + ylab("m (g)") + ggtitle("Nombres y unidades en los Ejes")
```

```r
#modelo
r <- seq(0, max(d$r)*1.1, length.out = 1000);
mmax <- 4*pi*r^3 * max(d$rho) / 3;
mmin <- 4*pi*r^3 * min(d$rho) / 3;

modelo_d3 <- data.frame(m_max = mmax, m_min = mmin, r = r);
```

```r
#ploteamos modelo
gmodelo <- ggplot() + 
geom_point(data = d, aes(x = r, y = m, colour = Nombre)) + 
geom_smooth(data = d, aes(x = r, y = m, colour = Nombre), method = "lm", formula = "y ~ x") + 
#geom_line(data = modelo_d3, aes(x = r, y = m_min, colour = "3d")) + 
#scale_colour_manual(values = c("Datos" = "Black", "3d" = "Black"),
#                    labels = list(TeX("Datos Experimentales"), TeX("$m \\propto \\frac{4 \\pi}{3}r^3$") ) ) + 
xlab("r (cm)") + ylab("m (g)") + ggtitle("Nombres y unidades en los Ejes") + temajuan + 
scale_y_continuous(trans = "log10") + 
scale_x_continuous(trans = "log10")  

show(gmodelo)
```

```r
Df  <- vector(mode = "numeric")
sDf <- vector(mode = "numeric")
Nombre <- vector(mode = "character")

for(i in unique(d$Nombre)){
di <- d[d$Nombre == i, ]
    
a <- summary.lm(lm(log10(di$m) ~ log10(di$r)))
Df <- c(Df,a$coefficients[2,1])
sDf <- c(sDf, a$coefficients[2,2])    
Nombre <- c(Nombre, i) 
}

resumen <- data.frame(Nombre = Nombre, Df = Df, sDf = sDf)
head(resumen, 14)
```
