a) Calcule la densidad de un esfera sólida de masa m = ( 10.0 ± 0.2 ) kg y radio r = ( 0.650 ±
0.001 ) dm (1dm = 10cm).

Solución:

$$\rho = \frac{m}{V} = \frac{m}{(4/3)\pi r^3}$$

$$\delta \rho = \Big|\frac{1}{(4/3)\pi r^3} \Big| \delta m + \Big| \frac{m}{(4\, /\, 9)\pi r^4} \Big| \delta r$$

```r
rho = 10 / (4/3 * pi * 0.65^3);
d_rho = (4/3 * pi * 0.65^3)^(-1) * 0.2 + ( 10 / ( (4/9) * pi * 0.65^4) ) * 0.001;
cat("\nLa densidad es\n\nrho +- drho = (",  rho," +- ", d_rho, ") kg / dm ^3")
```
