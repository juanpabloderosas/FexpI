```r
x <- seq(1,10,1);
y1 <- x *3 + 1 + rnorm(mean = 0, sd = 0.5, n = length(x));
y2 <- x *3 + 1 + rnorm(mean = 0, sd = 2, n = length(x))
datos <- data.frame(rep(x,2), c(y1,y2), c(rep("y1", length(x)),rep("y2", length(x))))
colnames(datos) <- c("x", "y", "orig")
print(datos)
```

```r
library("ggplot2")
g <- ggplot(data = datos) + geom_point(aes(x = x, y = exp(-x/2), colour = orig))
show(g)
```

```r
library("plotly");library("htmlwidgets")
ggplotly(g) -> gp
setwd("/home/juan/Escritorio/")
saveWidget(as_widget(gp), "ricardo.html")
```
